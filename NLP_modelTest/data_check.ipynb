## 02. VOD_MART 데이터프레임 불러오기

```python
import csv
```

```python
f = open('../data/vod_mart_data.csv', 'rt', encoding='utf8')
reader = csv.reader(f)

csv_list = []
for l in reader:
    csv_list.append(l)
f.close()
```

```python
vod_mart = pd.DataFrame(csv_list)
vod_mart.head()
```

```python
vod_mart.columns
```

```python
# 파일 전체 스캔하면서 Row별 Column	 개수 세기
delm = ','
with open('../data/vod_mart_data.csv', "r", encoding='utf8') as f:
	count_columns = [len(l.split(delm)) for l in f.readlines()]
# 가장 큰 Column 개수 기준 Columns명 리스트 만들어주기 
columns = [i for i in range(0, max(count_columns))]

df = pd.read_csv("../data/vod_mart_data.csv", names=columns, delimiter=',')
```
